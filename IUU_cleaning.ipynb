```python
import pandas as pd
import os
```

```python
data_2012 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2012_iuu_insights_v1_pipe4.csv')
data_2013 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2013_iuu_insights_v1_pipe4.csv')
data_2014 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2014_iuu_insights_v1_pipe4.csv')
data_2015 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2015_iuu_insights_v1_pipe4.csv')
data_2016 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2016_iuu_insights_v1_pipe4.csv')
data_2017 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2017_iuu_insights_v1_pipe4.csv')
data_2018 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2018_iuu_insights_v1_pipe4.csv')
data_2019 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2019_iuu_insights_v1_pipe4.csv')
data_2020 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2020_iuu_insights_v1_pipe4.csv')
data_2021 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2021_iuu_insights_v1_pipe4.csv')
data_2022 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2022_iuu_insights_v1_pipe4.csv')
data_2023 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2023_iuu_insights_v1_pipe4.csv')
data_2024 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2024_iuu_insights_v1_pipe4.csv')
data_2025 = pd.read_csv(r'E:\documents\DataScience\practicum\IUU_Regions\regions_2025_iuu_insights_v1_pipe4.csv')
```

```python
data_2012 = data_2012[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2012['layer'].notnull()]
data_2013 = data_2013[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2013['layer'].notnull()]
data_2014 = data_2014[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2014['layer'].notnull()]
data_2015 = data_2015[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2015['layer'].notnull()]
data_2016 = data_2016[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2016['layer'].notnull()]
data_2017 = data_2017[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2017['layer'].notnull()]
data_2018 = data_2018[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2018['layer'].notnull()]
data_2019 = data_2019[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2019['layer'].notnull()]
data_2020 = data_2020[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2020['layer'].notnull()]
data_2021 = data_2021[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2021['layer'].notnull()]
data_2022 = data_2022[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2022['layer'].notnull()]
data_2023 = data_2023[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2023['layer'].notnull()]
data_2024 = data_2024[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2024['layer'].notnull()]
data_2025 = data_2025[['mmsi','year','hours','active_hours','previously_iuu_listed','iuu_listed','layer']][data_2025['layer'].notnull()]
```

```python
df_2012 = data_2012.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2013 = data_2013.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2014 = data_2014.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2015 = data_2015.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2016 = data_2016.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2017 = data_2017.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2018 = data_2018.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2019 = data_2019.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2020 = data_2020.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2021 = data_2021.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2022 = data_2022.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2023 = data_2023.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2024 = data_2024.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
df_2025 = data_2025.groupby('layer').agg({'hours':'sum','active_hours':'sum','mmsi':'count','year':'first'}).reset_index()
```

```python
merge_df = pd.concat([df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024, df_2025], axis=0)
merge_df['year'] = merge_df['year']
merge_df.rename(columns={'mmsi':'mmsi_count'}, inplace=True)
merge_df
```

```python
output_df = pd.read_excel(r'E:\documents\DataScience\practicum\output.xlsx')
output_df
```

```python
final_df = pd.merge(merge_df, output_df, how='right', left_on=['layer','year'], right_on=['region','year'])
final_df.drop(['layer'], axis=1, inplace=True)
final_df
```

```python
final_df.to_csv(r'E:\documents\DataScience\practicum\IUU_outputWith2012.csv', index=False)
```
